# MRW

```python
df_mrw = pd.read_csv("./MRWCleanedV2.csv")
districts_df = pd.read_json('./districts.json', orient='records')
provinces_df = pd.read_json('./provinces.json', orient='records')
```

```python
merge_geo_df = pd.merge(
    districts_df,
    provinces_df,
    how='left',
    on='provinceCode',
)
```

```python
df_mrw = df_mrw[['UserId', 'ProductSKU', 'Distributor', 'ProductName',
       'ScanDate', 'Count', 'Points', 'Volume (L)', 'Full name', 'ProductLine',
       'UserIdOwner', 'GarageId', 'UserType', 'Postcode', 'Province',
       'District']].copy()
```

```python
df_mrw.columns = ['UserId', 'ProductSKU', 'Distributor', 'ProductName',
       'ScanDate', 'Count', 'Points', 'Volume', 'Fullname', 'ProductLine',
       'UserIdOwner', 'GarageId', 'UserType', 'Postcode', 'Province',
       'District']
```

```python
df_mrw['Postcode'] = df_mrw['Postcode'].astype('Int64').astype('string')
merge_geo_df['postalCode'] = merge_geo_df['postalCode'].astype('string')
```

```python
def rename_provice(provice):
        return merge_geo_df[merge_geo_df['provinceNameTh'] == provice]['provinceNameEn'].to_list()[0]
```

```python
df_mrw.loc[df_mrw['Postcode'].isna(),['Province']] = df_mrw.loc[df_mrw['Postcode'].isna(),['Province']]['Province'].apply(rename_provice)
df_mrw.loc[~df_mrw['Postcode'].isna(),['Province']] = df_mrw.loc[~df_mrw['Postcode'].isna(),['Province']]['Province'].apply(rename_provice)
```

```python
df_mrw_new = df_mrw[['UserId', 'ProductSKU', 'Distributor', 'ProductName',
       'ScanDate', 'Count', 'Points', 'Volume', 'Fullname', 'ProductLine',
       'UserIdOwner', 'GarageId', 'UserType', 'Postcode', 'Province', 'District']].copy()

df_mrw_new.columns = ['UserId', 'ProductSKU', 'Distributor', 'ProductName',
       'ScanDate', 'Count', 'Points', 'Volume', 'Fullname', 'ProductLine',
       'UserIdOwner', 'GarageId', 'UserType', 'Postcode', 'Province',
       'District']
```
